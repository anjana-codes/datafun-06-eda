7. Initial Visulaizations

Goal 1. Which are the states with highest number of crashes and lowest number of crashes?

 A Bar plot: A bar plot is used to visualize the states with highest number of crashes and lowest number of crashes.


```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))

# Load the car crash dataset
df = sns.load_dataset("car_crashes").sort_values("total", ascending=False)

# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(x="total", y="abbrev", data=df,
            label="Total", color="b")
```

Result: South Carolina, Noth Dakota and West Virginia have the highest number of crahes. Washington DC, Massachussets and Minnesota have the lowest number of crahes among all states.

Goal 2.  Is there any linear relationship betweeen total crashes and insurance premium?

 A Scatter Plot: A scatter plot is used to visulaize the relationship between total crashes and  insurance premiums. I used trend line to check whether the relationship is positive, negative or not    significant.

```python

# Extract the relevant columns
total_crashes = df['total']
insurance_premiums = df['ins_premium']

# Calculate the linear regression line
slope, intercept = np.polyfit(total_crashes, insurance_premiums, 1)
trend_line = slope * total_crashes + intercept

# Create scatter plot with trend line
plt.figure(figsize=(10, 6))
plt.scatter(total_crashes, insurance_premiums, color='blue', alpha=0.6, label='Data points')
plt.plot(total_crashes, trend_line, color='red', linewidth=2, label='Trend line')
plt.title('Scatter Plot: Total Crashes vs. Insurance Premiums with Trend Line')
plt.xlabel('Total Crashes')
plt.ylabel('Insurance Premiums')
plt.legend()
plt.grid(True)
plt.show()
```

Result: The scatter plot shows that the relationship isnot strong between toltal crashes and insurance premium. I also calculated the slope relationship and drew the trend line. The trend line shows a slightly negative relationship between total crashes and insurance premiums. These relationship hasn't been tested for statistical significance, though.

Goal 3. What is the relationship between different variables?

Correlation heatmap: A correlation heatmap is used to see relationships between different variables.

```python
# Drop the 'abbrev' column as it's not numerical
data = df.drop(columns='abbrev')

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)

# Display the heatmap
plt.title('Correlation Heatmap of Car_crashes Data')
plt.show()
```

Result: The correlation heatmap shows that the variables are mostly positively correlated each other. no_previous(proportion of drivers involved in fatal crashes with no previous incidents) variable has the strongest corelation (r=0.96) with the total crashes. Similary, not_distarcted (proportion of drivers involved in fatal crashes who were not distracted) and alcohol (proportion of drivers involved in fatal crashes who were alcohol-impaired) variables are also positively corelated with total crashes with r value 0.83 and 0.85 respectively. ins_losses (average insurance losses in the state per insured driver) and ins_premium (average insurance premium in the state) variables do not seem to exhibhit significant corelation with total crashes.
